```python
def PreProcess(data):
               
    # Preprocess feature columns
    outX = pd.DataFrame(index=data.index)  # output dataframe, initially empty

    # Check each column
    for col, col_data in data.iteritems():
        if (col_data.dtype == object):
            
            # Note: This change the 2 categorical A/B columns data types to int 0,1  
            if len(data[col].unique()) == 2:
                    col_data = col_data.replace(['A', 'B'], [1, 0])
                 
            # If non-numeric, convert to one or more dummy variables
            elif len(data[col].unique()) > 2:
                col_data = pd.get_dummies(col_data, prefix=col)  # e.g. 'action_type' => 'action_type_Jump Shot', 
                                                                 #'action_type_Driving Dunk Shot'
            
        outX = outX.join(col_data)  # collect column(s) in output dataframe
    
    outX = outX.dropna()  

    return outX 
```

```python
# Extract feature (X) and target (y) columns
feature_cols = list(processed_claims_train_data.columns[1:-1])  # all columns but first(id) and last(loss) are features
target_col = processed_claims_train_data.columns[-1]  # last column is the target/label

X_all = processed_claims_train_data[feature_cols]  # feature values for all students
y_all = processed_claims_train_data[target_col]  # corresponding targets/labels
```

```python
from sklearn.utils import shuffle
from sklearn.cross_validation import train_test_split

def shuffle_split_data(X, y):
    """ Shuffles and splits data into 70% training and 30% testing subsets,
        then returns the training and testing subsets. """
    
    # Shuffle and split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

    # Return the training and testing data subsets
    return X_train, X_test, y_train, y_test
```

```python
# Test shuffle_split_data
try:
    X_train, X_test, y_train, y_test = shuffle_split_data(X_all, y_all)
    print "Successfully shuffled and split the data!"
except:
    print "Something went wrong with shuffling and splitting the data."
```

```python
# Put any import statements you need for this code block here
from sklearn import metrics

def performance_metric(y_true, y_predict):
    """ Calculates and returns the total error between true and predicted values
        based on a performance metric chosen by the student. """
    
    return metrics.mean_absolute_error(y_true, y_predict) #evaluate performance 
```

```python
# Put any import statements you need for this code block
from sklearn.grid_search import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

def fit_DTRmodel(X, y):
    """ Tunes a decision tree regressor model using GridSearchCV on the input data X 
        and target labels y and returns this optimal model. """

     # Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

     # Set up the parameters we wish to tune
    parameters = {'max_depth':(20,30),
                  'min_samples_split':(300,500)}

    # Make the GridSearchCV object
    reg = GridSearchCV(regressor, param_grid = parameters)    
    reg.fit(X,y)  
    
    # Return the optimal model
    return reg


# Let's calculaate the error on our unseen data
try:
    start = time.time()
    reg = fit_DTRmodel(X_train, y_train)
    end = time.time()
    print "Done!\nTraining time (secs): {:.3f}".format(end - start)
    pred = reg.predict(X_test)
    total_error = performance_metric(y_test, pred)
    print "Successfully performed a metric calculation!"
    print "Total_error = {}".format(total_error)   
except:
    print "Something went wrong with fitting a model."
```

```python
import xgboost as xgb

def fit_XGBmodel(X,y):

    xgdmat = xgb.DMatrix(X.values, y.values) # Create our DMatrix to make XGBoost more efficient

    params = {'eta': 0.01, 'seed':0, 'subsample': 0.5, 'colsample_bytree': 0.5, 
             'objective': 'reg:linear', 'max_depth':6, 'min_child_weight':3} 
    
    num_rounds = 1000
    bst = xgb.train(params, xgdmat, num_boost_round = num_rounds)
    return bst

# Test fit_model on entire dataset
try:
    start = time.time()
    bst = fit_XGBmodel(X_train, y_train)
    end = time.time()
    print "Done!\nTraining time (secs): {:.3f}".format(end - start)
    test_xgb = xgb.DMatrix(X_test.values)
    pred = bst.predict(test_xgb)
    total_error = performance_metric(y_test, pred)
    print "Successfully performed a metric calculation!"
    print "Total_error = {}".format(total_error)   
except:
    print "Something went wrong with fitting a model."
```
